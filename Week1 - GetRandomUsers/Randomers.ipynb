### Project Description

The main task in this project is to connect to the Random User API, https://randomuser.me/ , and fetch randomly generated user data. 

After fetching the data, filter it and generate a list of the top 100 male users.

## Project Code & Data

```python
## Import Necessary Python Libraries

import pandas as pd
import numpy as np
from pandas import json_normalize
import requests
import json
```

```python
# Connect to the Random User API
url = "https://randomuser.me/api/?results=5000"
res_API = requests.get(url)

#data = response.json()
#data = json.loads(res_API.text)

# Get the data from the API
# Parse the data into JSON format
data = res_API.text
json.loads(data)
```

```python
# Create a file and open it in write mode
with open("data.json", "w") as write_file:
    json.dump(data, write_file)
```

```python
# Convert json to dataframe using json_normalize()
dict = json.loads(data)
data_df = json_normalize(dict['results'])

data_df.head(10)
```

```python
# Check the number of columns and rows
data_df.shape
```

```python
# Check the structure of the dataset
data_df.info()
```

```python
# Create a new dataframe with filtered data by male gender
male_users = data_df[data_df['gender'] == 'male']

male_users[:10]
```

```python
# Create a new dataframe and sort the data by index in ascending order
# Limit the number of rows to the top 100 rows
top_100musers = male_users.sort_index(ascending=True)[:100]
top_100musers[:10]
```

```python
# Check the shape of the top_100musers
top_100musers.shape
```

```python
# Save the top_100musers in csv form
top_100musers.to_csv('topmusers.csv')
```
